```python
import numpy as np
import matplotlib.pyplot  as plt
import cv2
```

```python
img1 = cv2.imread("1.jpg")
img = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
```

```python
edges = cv2.Canny(img, 50,150, apertureSize = 3)
```

```python
plt.imshow(edges, cmap = "gray")
```

**For Lines**

```python
lines = cv2.HoughLines(edges,1,np.pi/180,200)
```

```python
for line in lines:
    rho,theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho
    y0 = b * rho
    
    
    x1 = int(x0 + 1000 * (-b))  
    y1 = int(y0 + 1000 * (a))    
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * (a))
    cv2.line(img1,(x1,y1),(x2,y2),(0,0,255),1)
    
```

```python
plt.imshow(img1)
```

```python
lines = cv2.HoughLinesP(edges,1,np.pi/180,100,minLineLength=100,maxLineGap=10)
```

```python
for line in lines:
     x1,y1,x2,y2 = line[0]
     cv2.line(img,(x1,y1),(x2,y2),(0,255,0),1)
```

```python
plt.imshow(img)
```

**For Circles**

```python
img2 = cv2.imread('shapes.png')
```

```python
img3 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
edges1 = cv2.Canny(img3, 50, 150, apertureSize=3)
```

```python
plt.imshow(img3, cmap="gray")
plt.show()
```

```python
img4 = cv2.medianBlur(img3, 5)
```

```python
circles = cv2.HoughCircles(img3, cv2.HOUGH_GRADIENT, 1, 20, param1=50, param2=30, minRadius=0, maxRadius=0)
```

```python
circles
```

```python
detected_circles = np.uint16(np.around(circles))
for (x, y, r) in detected_circles[0, :]:
    cv2.circle(img2, (x, y), r, (255, 255, 255), 3)
    cv2.circle(img2, (x, y), 2, (0, 255, 255), 1)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)) 
plt.show()
```
