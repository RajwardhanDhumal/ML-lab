```python
def MSE(actual,predictions):
    actual = np.array(actual)
    predictions = np.array(predictions)
    squared_diff = (actual - predictions) ** 2
    mse = np.mean(squared_diff)
    return mse
```

```python
def rmse(y_true, y_pred):
   
    rmse = np.mean((y_true - y_pred) ** 2)
    return np.sqrt(rmse)


rmse_train = rmse(y_train, y_train_pred)
rmse_test = rmse(y_test, y_test_pred)

print("RMSE (Training):", rmse_train)
print("RMSE (Testing):", rmse_test)
```

```python
def cal_sigma(data):
    data = data.to_numpy()
    mean_data = data.mean()
    sigma_squared = (data - mean_data) ** 2
    sigma_squared_mean = sigma_squared.mean()
    return np.sqrt(sigma_squared_mean)
```

```python
sigma_of_columns = {c: i for i, c in enumerate(data.columns)}
sigma_of_columns
```

```python
for col in list(data):
    sigma_of_columns[col] = cal_sigma(data[col])
```
