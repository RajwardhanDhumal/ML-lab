```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import cvlib as cv # pip install cvlib
from cvlib.object_detection import draw_bbox
```

```python
im = cv2.imread('A1 Images/p2.jpeg')

plt.imshow(im[:,:,::-1])

bbox , conf = cv.detect_face(im)
```
