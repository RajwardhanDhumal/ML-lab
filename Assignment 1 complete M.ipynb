```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def rajwardhan_Translate(tx,ty,src_img,img_type):
    if img_type==0:
        h,w=src_img.shape
        T = np.float32([[1, 0, tx], [0, 1, ty]])
        I_T = cv2.warpAffine(src_img, T, (w,h))
    if img_type==1:
        h,w,_=src_img.shape
        T = np.float32([[1, 0, tx], [0, 1, ty]])
        I_T = cv2.warpAffine(src_img, T, (w,h))
    return I_T
```

```python
def rajwardhan_Rotate(an,src_img,sf,img_type):
    if img_type==0:
        h,w=src_img.shape
        cr=(w/2,h/2)
        M = cv2.getRotationMatrix2D(cr,an,sf)
        I_Rot = cv2.warpAffine(src_img,M,(w,h))
    if img_type==1:
        h,w,_=src_img.shape
        cr=(w/2,h/2)
        M=cv2.getRotationMatrix2D(cr,an,sf)
        I_Rot=cv2.warpAffine(src_img,M,(w,h))   
    return I_Rot
```

```python
def rajwardhan_Scaling(sx, sy, src_img, img_type):
    if img_type == 0:
        I_S = cv2.resize(src_img, dsize=None, fx=sx, fy=sy)
    if img_type == 1:
        I_S = cv2.resize(src_img, dsize=None, fx=sx, fy=sy) 
    return I_S
```

```python
def rajwardhan_Reflection(axis,src_img,img_type):
    if img_type==0:
        h,w=src_img.shape
        if axis==0:
            Rx=np.float32([[1,0,0],[0,-1,h]])
            I_R=cv2.warpAffine(src_img,Rx,(w,h))
        if axis==1:
            Ry=np.float32([[-1,0,w],[0,1,0]])
            I_R=cv2.warpAffine(src_img,Ry,(w,h))
    if img_type==1:
        h,w,_=src_img.shape
        if axis==0:
            Rx=np.float32([[1,0,0],[0,-1,h]])
            I_R=cv2.warpAffine(src_img,Rx,(w,h))
        if axis==1:
            Ry=np.float32([[-1,0,w],[0,1,0]])
            I_R=cv2.warpAffine(src_img,Ry,(w,h))
    return I_R
```

```python
def rajwardhan_Shearing(axis,src_img,img_type):
    if img_type==0:
        h,w = src_img.shape
    elif img_type==1:
        h,w,_ = src_img.shape
    if axis==0:
        shx=float(input("Enter Shx"))
        SHx=np.float32([[1,shx,0],[0,1,0]])
        SHx_I=cv2.warpAffine(src_img,SHx,(w,h))
        return SHx_I
    if axis==1:
        shy=float(input("Enter Shy"))
        SHy=np.float32([[1,0,0],[shy,1,0]])
        SHy_I=cv2.warpAffine(src_img,SHy,(w,h))
        return SHy_I
```

```python
def rajwardhan_Switch_1(ch,src_img,img_type):
    axis=0
    if ch==1:
        tx=int(input("Enter tx:"))
        ty=int(input("Enter ty:"))
        img=rajwardhan_Translate(tx,ty,src_img,img_type)
        return img
    elif ch==2:
        an=int(input("Enter Angle:"))
        sf=int(input("Enter Scaling factor:"))
        img=rajwardhan_Rotate(an,src_img,sf,img_type)
        return img
    elif ch==3:
        sx=float(input("Enter sx:"))
        sy=float(input("Enter sy:"))
        img=rajwardhan_Scaling(sx,sy,src_img,img_type)
        return img
    elif ch==4:
        axis=int(input("Enter 0 for x and 1 for y:"))
        img=rajwardhan_Reflection(axis,src_img,img_type)
        return img
    elif ch==5:
        axis=float(input("Enter 0 for Shx and 1 for Shy:"))
        img=rajwardhan_Shearing(axis,src_img,img_type)
        return img
    else:
        print("Wrong choice")
        return 0
```

```python
def rajwardhan_Switch_0(ch):
    src_path=input("Enter Image name:")
    ch1=int(input("Enter 1 for Translation \nEnter 2 for Rotation \nEnter 3 for Scaling \nEnter 4 for Reflection \nEnter 5 for Shearing"))
    if ch==1:
        img_src=cv2.imread(src_path,cv2.IMREAD_GRAYSCALE)
        img=rajwardhan_Switch_1(ch1,img_src,0)
        return img
    elif ch==2:
        img_src=cv2.imread(src_path)
        img=rajwardhan_Switch_1(ch1,img_src,1)
        return img
    else:
        print("Wrong choice")
        return 0
        
        
```

```python
def Main():
    ch=int(input("Enter 1 for B/W \nEnter 2 for Color\n"))
    img=rajwardhan_Switch_0(ch)
    if ch==1:
        plt.imshow(img,cmap="gray")
    if ch==2:
        img=img[:,:,::-1]
        plt.imshow(img)
    
```
