```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r_ones = np.ones((100, 100), dtype = int) 
```

#one means only one in matrix 


```python
one_two = np.c_[r_ones, 2*r_ones]
```

#In NumPy, .c_ is a special object that is used to concatenate arrays along the second axis, effectively adding columns. It is a simple way to create a new array by stacking existing arrays horizontally.

```python
three_four = np.c_[3*r_ones, 4*r_ones]
```

```python
linear_t = np.vstack([three_four, one_two])
```

```python
plt.imshow(linear_t)
```

```python
old_x = 100
scaling_factor = 1.55
new_x = int(old_x * scaling_factor)
new_x
```

```python
linear_t[0:new_x, 0:100] = 3
```

```python
plt.imshow(linear_t)
```

### Hence, we are done with scaling this quadrant of 1s with a scaling factor

```python
old_x = 100
scaling_factor = 2.0
new_x = int(old_x * scaling_factor)  # new_x will be 200


linear_t = np.zeros((new_x, 150))  # Create a 200x150 array of zeros

linear_t[0:new_x, 0:100] = 4  # Set the region from rows 0 to 200 and columns 0 to 100 to the value 4

plt.imshow(linear_t, cmap='viridis')  # 'viridis' is a colormap; you can choose any colormap you prefer
plt.colorbar()  # Show color scale
plt.title('Visualization of linear_t')
plt.show()
```
